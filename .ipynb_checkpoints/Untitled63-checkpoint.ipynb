```python
import cv2
import numpy as np

# 画像の読み込み
image = cv2.imread('image1.jpeg')

# 画像をHSV色空間に変換
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# 茶色の領域の範囲を定義
lower_brown = np.array([10, 0, 100])  # 茶色の下限(HSV)
upper_brown = np.array([30, 255, 170])  # 茶色の上限(HSV)

# 茶色の領域を抽出するマスクを作成
brown_mask = cv2.inRange(hsv_image, lower_brown, upper_brown)

# マスクを使って元の画像から茶色の領域を抽出
brown_area = cv2.bitwise_and(image, image, mask=brown_mask)

# 結果の表示
cv2.imshow('Original Image', image)
cv2.imshow('Brown Area', brown_area)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

# 画像の読み込み
image = cv2.imread("image1.jpeg")
cv2.imshow("image", image)
# 画像をHSV色空間に変換
hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# 黒色の範囲を定義
lower_black = np.array([10, 0, 0])
upper_black = np.array([200, 200, 155])

# 黒色領域のマスクを作成
mask = cv2.inRange(hsv, lower_black, upper_black)

# マスクを使って黒色領域を抜き出す
result = cv2.bitwise_and(image, image, mask=mask)

# 結果を表示
cv2.imshow("Result", result)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

# 画像の読み込み
image = cv2.imread('image1.jpeg')

# 画像をHSV色空間に変換
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# 茶色の領域の範囲を定義
lower_brown = np.array([10, 0, 50])  # 茶色の下限(HSV)
upper_brown = np.array([200, 200, 155])  # 茶色の上限(HSV)

# 茶色の領域を抽出するマスクを作成
brown_mask = cv2.inRange(hsv_image, lower_brown, upper_brown)

# マスクを使って元の画像から茶色の領域を抽出
brown_area = cv2.bitwise_and(image, image, mask=brown_mask)

# 輪郭を検出
contours, _ = cv2.findContours(brown_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# 各輪郭に対して領域を囲む矩形を描画
for contour in contours:
    x, y, w, h = cv2.boundingRect(contour)
    cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)

# 結果の表示
cv2.imshow('Original Image', image)
cv2.imshow('Brown Area', brown_area)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#黒取り出す
import cv2
import numpy as np

# 画像の読み込み
image = cv2.imread('1_1.jpg')

# 画像をHSV色空間に変換
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# 黒色の領域の範囲を定義
lower_brown = np.array([0, 0, 50])  # 茶色の下限(HSV)
upper_brown = np.array([255, 200, 155])  # 茶色の上限(HSV)

# 茶色の領域を抽出するマスクを作成
brown_mask = cv2.inRange(hsv_image, lower_brown, upper_brown)

# マスクを使って元の画像から茶色の領域を抽出
brown_area = cv2.bitwise_and(image, image, mask=brown_mask)

# 輪郭を検出
contours, _ = cv2.findContours(brown_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# 各輪郭に対して領域を囲む矩形を描画
for contour in contours:
    # 輪郭の領域を計算
    area = cv2.contourArea(contour)

    # 領域が1mm以上の場合に矩形を描画
    if  7.5 < area < 20:
        x, y, w, h = cv2.boundingRect(contour)
        cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)

# 結果の表示
cv2.imshow('Original Image', image)
cv2.imshow('Brown Area', brown_area)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#白の輪郭内部から黒を取り出す

import cv2
import numpy as np

# 画像の読み込み
image = cv2.imread('image1.jpeg')
cv2.imshow('image', image)
# 画像をグレースケールに変換
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2.imshow('gray_image', gray_image)

# 白い領域のしきい値範囲を設定
lower_white = 200
upper_white = 255

# しきい値処理
_, binary_image = cv2.threshold(gray_image, lower_white, upper_white, cv2.THRESH_BINARY)

# 輪郭の検出
contours, _ = cv2.findContours(binary_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# 面積が最大の輪郭を見つける
max_contour = max(contours, key=cv2.contourArea)

# 最大の輪郭を囲む四角を取得
x, y, w, h = cv2.boundingRect(max_contour)

# 四角で最大の輪郭を囲む
cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)
cv2.imshow('_image', image)
# 指定した領域を切り抜く
cropped_image = image[y:y+h, x:x+w]
cv2.imshow('cropped_image', cropped_image)

# 画像をHSV色空間に変換
hsv_image = cv2.cvtColor(cropped_image, cv2.COLOR_BGR2HSV)

# 茶色の領域の範囲を定義
lower_brown = np.array([0, 0, 50])  # 茶色の下限(HSV)
upper_brown = np.array([255, 200, 155])  # 茶色の上限(HSV)

# 茶色の領域を抽出するマスクを作成
brown_mask = cv2.inRange(hsv_image, lower_brown, upper_brown)

# マスクを使って元の画像から茶色の領域を抽出
brown_area = cv2.bitwise_and(cropped_image, cropped_image, mask=brown_mask)

# 輪郭を検出
contours, _ = cv2.findContours(brown_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# 各輪郭に対して領域を囲む矩形を描画
for contour in contours:
    # 輪郭の領域を計算
    area = cv2.contourArea(contour)

    # 領域が1mm以上の場合に矩形を描画
    if  8 < area < 20:
        x, y, w, h = cv2.boundingRect(contour)
        cv2.rectangle(cropped_image, (x, y), (x + w, y + h), (0, 255, 0), 2)

# 結果の表示
cv2.imshow('Original Image', cropped_image)
cv2.imshow('Brown Area', brown_area)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

# 画像を読み込む
image = cv2.imread('1_1.jpg')

# 画像のサイズを1/4に縮小する
resized_image = cv2.resize(image, (0, 0), fx=0.25, fy=0.25)

# 画像をHSV色空間に変換
hsv_image = cv2.cvtColor(resized_image, cv2.COLOR_BGR2HSV)

# 黒色の領域の範囲を定義
lower_brown = np.array([0, 0, 50])  # 茶色の下限(HSV)
upper_brown = np.array([255, 200, 155])  # 茶色の上限(HSV)

# 茶色の領域を抽出するマスクを作成
brown_mask = cv2.inRange(hsv_image, lower_brown, upper_brown)

# マスクを使って元の画像から茶色の領域を抽出
brown_area = cv2.bitwise_and(image, image, mask=brown_mask)

# 輪郭を検出
contours, _ = cv2.findContours(brown_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# 各輪郭に対して領域を囲む矩形を描画
for contour in contours:
    # 輪郭の領域を計算
    area = cv2.contourArea(contour)

    # 領域が1mm以上の場合に矩形を描画
    if  7.5 < area < 20:
        x, y, w, h = cv2.boundingRect(contour)
        cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)

# 結果の表示
cv2.imshow('Original Image', image)
cv2.imshow('Brown Area', brown_area)
cv2.waitKey(0)
cv2.destroyAllWindows()

# 結果の表示
cv2.imshow('Original Image', resized_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
