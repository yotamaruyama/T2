```python
import cv2

# 画像のパスを指定する
image_path = "konnbea4.jpeg"

# 画像を読み込む
image = cv2.imread(image_path)
cv2.imshow("image", image)

# グレースケールに変換する
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2.imshow("gray_image", gray_image)

# 2値化する
threshold_value = 150  # 閾値を適宜調整する
_, binary_image = cv2.threshold(gray_image, threshold_value, 255, cv2.THRESH_BINARY)

cv2.imshow("binary_image", binary_image)

# 輪郭を検出する
contours, _ = cv2.findContours(binary_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# 面積の範囲を指定して輪郭をフィルタリングする
min_contour_area = 1000  # 1mmの面積に対応する値（適宜調整する）
max_contour_area = 5000 # 5mmの面積に対応する値（適宜調整する）
filtered_contours = [
    cnt for cnt in contours if min_contour_area < cv2.contourArea(cnt) < max_contour_area
]

# 検出された輪郭に円を描画する
for contour in filtered_contours:
    (x, y, w, h) = cv2.boundingRect(contour)
    cv2.circle(image, (int(x + w/2), int(y + h/2)), 3, (0, 0, 255), -1)

# 画像を表示する
cv2.imshow("Detected Points", image)

# キー入力を待つ
cv2.waitKey(0)

# ウィンドウを閉じる
cv2.destroyAllWindows()
```

```python
import cv2

# 画像のパスを指定する
image_path = "konnbea4.jpeg"

# 画像を読み込む
image = cv2.imread(image_path)
cv2.imshow("image", image)

# グレースケールに変換する
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2.imshow("gray_image", gray_image)

# 2値化する
threshold_value = 50  # 閾値を適宜調整する
_, binary_image = cv2.threshold(gray_image, threshold_value, 255, cv2.THRESH_BINARY)
cv2.imshow("binary_image", binary_image)

# 輪郭を検出する
contours, _ = cv2.findContours(binary_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# 直径の範囲を指定して輪郭をフィルタリングする
min_diameter = 1  # 最小直径（単位：mm）
max_diameter = 5  # 最大直径（単位：mm）

filtered_contours = []
for contour in contours:
    (x, y, w, h) = cv2.boundingRect(contour)
    diameter = max(w, h)  # 輪郭の外接矩形の幅と高さのうち、大きい方が直径の近似となる

    # 直径が指定された範囲内であれば輪郭を追加する
    if min_diameter <= diameter <= max_diameter:
        filtered_contours.append(contour)
        cv2.rectangle(image, (x, y), (x + w, y + h), (0, 0, 255), 2)

# 画像を表示する
cv2.imshow("Detected Points", image)

# キー入力を待つ
cv2.waitKey(0)

# ウィンドウを閉じる
cv2.destroyAllWindows()
```
