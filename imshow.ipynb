```python
# 画像のダウンロード
import urllib.request as req
url = "https://uta.pw/shodou/img/28/214.png"
req.urlretrieve(url, "test.png")

# OpenCVで読み込む
import cv2
img = cv2.imread("test.png")
print(img)
```

```python
# ダウンロードした画像を画面に表示する
import matplotlib.pyplot as plt
import cv2
img = cv2.imread("test.png")
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```
